```python
# Print out bucket names
for bucket in s3.buckets.all():
    print(bucket.name)
```

## the US

```python
# parameters
buck_name = 'log.maymobility.com'
conn = client('s3')
bucket = s3.Bucket(buck_name)
#file path
parent_dir = '/Users/meredithmay/Downloads'
# import permission error logs' name
log_name = pd.read_csv(os.path.join(parent_dir,'data-1643035938872.csv'))
error = log_name[log_name.extracted_time.notna()].log_name
```

```python
def restore_permission_error(error_log_names):
    # specific file-name that needs restore
    f_list = ['log_info.json','log_stats.log','shift_info.json','state/push_info.json']
    for i in range(len(f_list)):
        f_list[i] =  error_log_names +'/'+ f_list[i] # other four files
    f_list.append(error_log_names +'/'+ error_log_names[-26:]+'.mcm')# mcm file
    
    # restore 
    for obj in f_list:
        try:
            print(f'file being restored {obj}')
            resp = bucket.meta.client.restore_object(
                Bucket=buck_name,
                Key= obj,
                RestoreRequest={'Days': 1, 'GlacierJobParameters': {'Tier': 'Bulk'}}
                )
        except:
            print(f'file not being restored {obj}')  
```

```python
for log_name in error:
    restore_permission_error(log_name)
```
